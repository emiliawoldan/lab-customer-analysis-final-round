### 04 - Processing Data

- Dealing with outliers.
- Normalization.
- Encoding Categorical Data.
- Splitting into train set and test set.

```python
cat = df_final.select_dtypes(include = np.object)
cat
```

```python
num = df_final.select_dtypes(include = np.number)
num
```

```python
def boxcox_transform(df_final):
    numeric_cols = df_final.select_dtypes(np.number).columns
    _ci = {column: None for column in numeric_cols}
    for column in numeric_cols:
        df_final[column] = np.where(df_final[column]<=0, np.NAN, df_final[column]) 
        df_final[column] = df_final[column].fillna(df_final[column].mean())
        transformed_data, ci = stats.boxcox(df_final[column])
        df_final[column] = transformed_data
        _ci[column] = [ci] 
    return df_final, _ci
```

```python
def remove_outliers(df_final, threshold=1.5, in_columns=df_final.select_dtypes(np.number).columns, skip_columns=[]):
    for column in in_columns:
        if column not in skip_columns:
            upper = np.percentile(df_final[column],75)
            lower = np.percentile(df_final[column],25)
            iqr = upper - lower
            upper_limit = upper + threshold * iqr
            lower_limit = lower - threshold * iqr
            df_final = df_final[(df_final[column]>lower_limit) & (df_final[column]<upper_limit)]
    return df_final
```

```python
df_final = pd.get_dummies(df_final, drop_first=True)
df_final
```

```python
df_final = remove_outliers(df_final, threshold=1.5, in_columns=['total_claim_amount'])
```

```python
X = df_final.drop('total_claim_amount', axis=1)
y = df_final['total_claim_amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### 05 - Modeling

- Apply model.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X) 
```

### 06 - Model Validation

- R2.
- MSE.
- RMSE.
- MAE.

```python
predictions = model.predict(X_test)
r2_score(y_test, predictions), mean_absolute_error(y_test, predictions), mean_squared_error(y_test, predictions, squared=False)
```
